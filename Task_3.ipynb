```python
import pandas as pd

# Load the dataset
iris = pd.read_csv('/content/ Iris.csv')
print(iris)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = 'darkgrid')
df = sns.load_dataset("iris")
fig, axs = plt.subplots(2, 2, figsize=(7,7))

sns.histplot(data=iris, x='SepalLengthCm', kde=True, color='skyblue', ax=axs[0,0])
sns.histplot(data=iris, x='SepalWidthCm', kde=True, color='olive', ax=axs[0,1])
sns.histplot(data=iris, x='PetalLengthCm', kde=True, color='gold', ax=axs[1,0])
sns.histplot(data=iris, x='PetalWidthCm', kde=True, color='teal', ax=axs[1,1])

plt.show()
```
